## 2. Load and check data
### 2.1 Let's create some synthetic data

```python
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding


X_swiss, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)
```

```python
X_swiss
```

```python


from matplotlib.colors import ListedColormap

darker_hot = ListedColormap(plt.cm.hot(np.linspace(0, 0.8, 256)))

axes = [-11.5, 14, -2, 23, -12, 15]

fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_swiss[:, 0], X_swiss[:, 1], X_swiss[:, 2], c=t, cmap=darker_hot)
ax.view_init(10, -70)


plt.show()
```

## 3. Let's apply t-SNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, init="random", learning_rate="auto",
            random_state=42)
X_reduced_tsne = tsne.fit_transform(X_swiss)
```

```python
plt.scatter(X_reduced_tsne[:, 0], X_reduced_tsne[:, 1],
            c=t, cmap=darker_hot)
plt.xlabel("$z_1$")
plt.ylabel("$z_2$", rotation=0)

plt.grid(True)


plt.title("Unrolled swiss roll using t-SNE")
plt.show()
```

#### Based on the previous figure we conclude that t-SNE was able to project the original dataset into a two dimensional projection quite well.
